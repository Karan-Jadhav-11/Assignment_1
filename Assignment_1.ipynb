9. Python supports the creation of anonymous functions at runtime, using a construct called 
Ans:In Python, the creation of anonymous functions at runtime is supported through a construct called "lambda" functions. Lambda functions are a way to create small, unnamed, and inline functions. The syntax for a lambda function is:
lambda arguments: expression
Here's an example of a simple lambda function that adds two numbers:
add = lambda x, y: x + y
result = add(3, 5)
print(result)  # Outputs: 8
In this example, lambda x, y: x + y defines an anonymous function that takes two arguments (x and y) and returns their sum. The lambda function is assigned to the variable add, and then it is called with the arguments 3 and 5, resulting in the output 8.
Lambda functions are often used for short, one-time operations, especially in cases where a full function definition is unnecessary.

14. How is memory managed in Python? 

ANS:Memory management in Python is primarily handled by the Python memory manager, which is part of the Python runtime environment. The memory manager is responsible for allocating memory for Python objects, managing the heap, and freeing up memory when objects are no longer needed.
Here are some key points to consider when it comes to memory management in Python:
1) Reference Counting: Each Python object has a reference count associated with it. This count is incremented every time a new reference is created to the object, and decremented every time a reference is destroyed. When the reference count reaches zero, the object is deallocated and its memory is freed.
Memory Allocation: The memory manager allocates memory in blocks called “pages.” When an object is created, the memory manager first checks 2) if there is enough free memory in the current page to allocate the object. If not, it allocates a new page. When an object is deallocated, its memory is returned to the current page, and can be reused for new objects.
Automatic Garbage Collection: In addition to reference counting, Python also includes a garbage collector that can identify and deallocate “cyclic” objects that are no longer reachable. A cyclic object is an object that has a reference to itself, either directly or indirectly, and would therefore never have its reference count reach zero under normal circumstances.
3) Memory Efficiency: Python optimizes memory utilization by allocating the same object reference to a new variable if the object already exists with the same value. For example, if you assign the integer 5 to two different variables, both variables will reference the same object in memory.

15. How to install Python on Windows and set path variables? 

Ans:Installing Python on Windows is a straightforward process, and setting the path variables allows you to run Python from any command prompt. Here are step-by-step instructions:
Installing Python:
Download Python:
Visit the official Python website: Python Downloads.
Choose the latest version for Windows (for example, Python 3.9.7).
Select the appropriate installer based on your system architecture (32-bit or 64-bit). Most modern systems are 64-bit.
Run the Installer:
Double-click the downloaded installer (e.g., python-3.9.7-amd64.exe) to start the installation.
Customize Installation (Optional):
During the installation, you'll have the option to customize the installation.
Ensure to check the box that says "Add Python 3.x to PATH" before proceeding.
You can also customize the installation location if needed.
Complete the Installation:
Click on the "Install Now" button to start the installation.
The installer will copy files and set up Python. Once completed, you'll see a screen indicating a successful installation.
Setting Path Variables:
After installing Python, you need to set the system PATH variable to include the directory where Python is installed. This allows you to run Python from any command prompt.
Find Python Installation Directory:
The default installation directory is usually C:\Users\YourUsername\AppData\Local\Programs\Python\Python3.x (replace YourUsername with your actual Windows username).
Copy Python Installation Directory Path:
Copy the path to the Python installation directory.
Set PATH Environment Variable:
Right-click on the Windows Start button and select "System."
Click on "Advanced system settings" on the left.
In the System Properties window, click on the "Environment Variables..." button.
Edit System PATH Variable:
In the Environment Variables window, under the "System variables" section, find and select the "Path" variable, then click on the "Edit..." button.
Add Python Directory to PATH:
Click on the "New" button and paste the path to the Python installation directory.
Verify Installation:
Open a new command prompt and type python or python --version to verify that Python is installed and the PATH variable is set correctly.

16. Is indentation required in python? 

Ans: Yes, indentation is required in Python and plays a crucial role in structuring the code. Unlike other programming languages where indentation is used for readability only, Python uses indentation to indicate a block of code. If you skip the indentation, Python will raise an error. The number of spaces for indentation is up to the programmer, but it has to be at least one space and consistent throughout the same block of code. Here’s an example to demonstrate the importance of indentation in Python:
if 5 > 2:
    print("Five is greater than two!")


In this example, the print statement is part of the if statement block because it is indented. If you don’t indent it, you will get an error.
